```python
def district_fitness(district_tracts, lambda_race=1.0):
    
    district_pop = merged.loc[district_tracts, "total_population"].sum()
    
    # Normalize population deviation
    pop_penalty = ((district_pop - target_pop) / target_pop) ** 2
    
    racial_penalty = 0
    
    for race in state_props.keys():
        district_race_pop = merged.loc[district_tracts, f"total_{race}"].sum()
        district_pct = district_race_pop / district_pop
        
        racial_penalty += (district_pct - state_props[race]) ** 2
    
    return pop_penalty + lambda_race * racial_penalty
```

```python
def map_fitness(district_map):
    # district_map: dict {district_id: [tract indices]}
    
    total_fitness = 0
    
    for d in district_map:
        total_fitness += district_fitness(district_map[d])
    
    return total_fitness
```

```python
import numpy as np
import random

def run_growth(neighbors, pop_array, target_pop, NUM_DISTRICTS=14):

    n_tracts = len(pop_array)
    district_assignment = np.full(n_tracts, -1)

    # Random seeds
    seeds = random.sample(range(n_tracts), NUM_DISTRICTS)

    for d, seed in enumerate(seeds):
        district_assignment[seed] = d

    # Grow districts
    while -1 in district_assignment:

        # Grow smallest population districts first
        district_pops = [
            pop_array[district_assignment == d].sum()
            for d in range(NUM_DISTRICTS)
        ]

        for d in np.argsort(district_pops):

            district_tracts = np.where(district_assignment == d)[0]

            candidate_tracts = []

            for tract in district_tracts:
                for neighbor in neighbors[tract]:
                    if district_assignment[neighbor] == -1:
                        candidate_tracts.append(neighbor)

            if not candidate_tracts:
                continue

            current_pop = pop_array[district_tracts].sum()

            best_candidate = None
            best_score = float("inf")

            for c in candidate_tracts:
                new_pop = current_pop + pop_array[c]
                score = abs(new_pop - target_pop)

                if score < best_score:
                    best_score = score
                    best_candidate = c

            if best_candidate is not None:
                district_assignment[best_candidate] = d

    return district_assignment
```

```python
def map_fitness(district_assignment,
                pop_array,
                race_matrix,
                target_pop,
                state_race_props,
                lambda_race=1.0,
                NUM_DISTRICTS=14):

    total_penalty = 0

    for d in range(NUM_DISTRICTS):

        mask = (district_assignment == d)

        district_pop = pop_array[mask].sum()

        # ---- Population penalty (normalized) ----
        pop_penalty = ((district_pop - target_pop) / target_pop) ** 2

        # ---- Racial penalty ----
        district_race_totals = race_matrix[mask].sum(axis=0)

        # Avoid divide-by-zero
        if district_pop > 0:
            district_race_props = district_race_totals / district_pop
        else:
            district_race_props = np.zeros_like(state_race_props)

        racial_penalty = np.sum(
            (district_race_props - state_race_props) ** 2
        )

        total_penalty += pop_penalty + lambda_race * racial_penalty

    return total_penalty
```

```python
import random
import networkx as nx

def mutate_map(district_assignment, neighbors, mutation_rate=0.01):

    new_map = district_assignment.copy()
    n_tracts = len(new_map)

    for tract in range(n_tracts):

        if random.random() < mutation_rate:

            current_district = new_map[tract]

            # Neighboring districts
            neighbor_districts = list({
                new_map[n] for n in neighbors[tract]
                if new_map[n] != current_district
            })

            if not neighbor_districts:
                continue

            proposed_district = random.choice(neighbor_districts)

            # Tentatively reassign
            new_map[tract] = proposed_district

            # Check contiguity of BOTH affected districts
            if not (is_contiguous(new_map, current_district, neighbors) and
                    is_contiguous(new_map, proposed_district, neighbors)):

                # revert
                new_map[tract] = current_district

    return new_map
```

```python
def is_contiguous(map_array, district_id, neighbors):

    tracts = np.where(map_array == district_id)[0]

    if len(tracts) == 0:
        return False

    G = nx.Graph()

    for t in tracts:
        for n in neighbors[t]:
            if n in tracts:
                G.add_edge(t, n)

    return nx.is_connected(G)
```

```python
POP_SIZE = 10
NUM_GENERATIONS = 30

population = []

# Initial population via growth
for _ in range(POP_SIZE):
    m = run_growth(neighbors, pop_array, target_pop)
    population.append(m)
```

```python
ELITE_SIZE = 3
mutation_rate = 0.01

for g in range(NUM_GENERATIONS):

    # Evaluate
    scores = [
        map_fitness(m, pop_array, race_matrix,
                    target_pop, state_race_props, lambda_race=1.0)
        for m in population
    ]

    # Sort by fitness (lower is better)
    sorted_indices = np.argsort(scores)
    population = [population[i] for i in sorted_indices]
    scores = [scores[i] for i in sorted_indices]

    print(f"Generation {g}: Best fitness = {scores[0]:.6f}")

    # Keep elites
    new_population = population[:ELITE_SIZE]

    # Fill rest with mutated versions
    while len(new_population) < POP_SIZE:

        parent = random.choice(population[:ELITE_SIZE])
        child = mutate_map(parent, neighbors, mutation_rate)
        new_population.append(child)

    population = new_population
```
