```python
import gymnasium as gym
import ale_py
from stable_baselines3 import DQN
from stable_baselines3.common.atari_wrappers import AtariWrapper

gym.register_envs(ale_py)

env = gym.make("ALE/KingKong-v5")
env = AtariWrapper(env)  # handles resizing, grayscale, frame stacking

model = DQN(
    "CnnPolicy",
    env,
    verbose=1,
    buffer_size=100000,
    learning_rate=1e-4,
    batch_size=32,
    train_freq=4,
    target_update_interval=1000,
)

model.learn(total_timesteps=100000)  # increase later

model.save("kingkong_dqn1")
```

```python
from stable_baselines3 import DQN

model = DQN.load("kingkong_dqn")
```

```python
env = gym.make("ALE/KingKong-v5", render_mode="human")
env = AtariWrapper(env)

obs, _ = env.reset()

while True:
    action, _ = model.predict(obs)
    obs, _, terminated, truncated, _ = env.step(action)

    if terminated or truncated:
        obs, _ = env.reset()
```
