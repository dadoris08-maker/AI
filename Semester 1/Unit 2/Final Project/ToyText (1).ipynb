# Given a 60% chance that the elf falls off of the cliff more than 150 times before the end of the round, what is the probability that the elf will end the round without falling off of the cliff more than 150 times for the first time on the 4th run?

In this simulation, we assume that the probability of falling off of the cliff more than 2000 times is 60%.
Each trial would be running the simulation until the elf reaches the goal. The observed result is how many times the elf falls off of the cliff, with the random variable being the probability.

```python
x=4
a=x-1
p=.40
q=1-p
tProb = (q**a)*p
print(tProb)
```

```python
import ale_py
import gymnasium as gym

env = gym.make("CliffWalking-v1", render_mode="ansi")  # 'ansi' gives text output
state, info = env.reset()
print("Initial state:", state)
```

```python
n_trials = 100
n_episodes = 5
count4 = 0  # Number of times success first occurs on 4th episode

for trial in range(n_trials):
    cliffCount = 0
    moveCount = 0
    success_found = False  # To track the first success in this trial

    for episode in range(n_episodes):
        state, info = env.reset()
        total_reward = 0
        done = False
        cliffCount = 0  # Reset for each episode

        while not done:
            action = env.action_space.sample()
            next_state, reward, terminated, truncated, info = env.step(action)

            if reward == -100:
                cliffCount += 1
            else:
                moveCount += 1

            total_reward += reward
            done = terminated or truncated

        print(f"Trial {trial + 1}, Episode {episode + 1}: Reward = {total_reward}, Cliff falls = {cliffCount}")

        # Check if this episode meets the condition (goal reached with <150 cliff falls)
        if cliffCount < 150 and not success_found:
            success_found = True  # first success found this trial
            if episode == 3:  # 4th episode (0-indexed)
                count4 += 1
                print(f"âœ… Success first occurred on episode 4 in trial {trial + 1}")

print(f"\nNumber of times success first occurred on episode 4: {count4}")
```

```python
p_empirical = count4 / n_trials
print(f"Empirical probability of first success on episode 4: {p_empirical:.4f}")
```

# Answer:

Despite assuming a probability of 60%, the empirical probability of successfully reaching the goal without falling off of the cliff 150 times on the fourth trial, 8%, is close to the theoretical probability, which is 8.63999%.
