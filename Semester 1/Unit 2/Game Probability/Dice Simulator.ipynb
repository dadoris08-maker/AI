```python
import random as rd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
rolls = []
numSides = 6
# Run the dice simulator 100 times
for i in range(100):
    roll = rd.randint(1, numSides)  # simulates rolling a 6-sided die
    rolls.append(roll)

# Print results
print("Number of sides:", numSides)
print("All dice rolls:", rolls)
print("Number of rolls:", len(rolls))
counts = Counter(rolls)

print("Dice roll counts:")
for number in range(1, numSides+1):
    print(f"{number}: {counts[number]}")
```

```python
# Plot histogram
plt.hist(rolls, bins=range(1, numSides+2), edgecolor="black", align="left")

# Add labels and title
plt.xlabel("Dice Value")
plt.ylabel("Frequency")
plt.title("Dice Rolls Simulation (100 rolls)")

plt.show()
```
