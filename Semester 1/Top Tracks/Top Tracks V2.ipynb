```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Added pandas import which was missing

df = pd.read_csv("DataTUSA.csv")
df["Genre_Over_1b"].value_counts().plot(kind="bar")

plt.xlabel("Genre")
plt.ylabel("Number of Songs with 1+ Billion Listens")
plt.title("Number of Songs with 1+ Billion Listens based on genre")
plt.show()
```

```python
df = pd.read_csv("DataTUSA.csv")
counts = df["Genre_Over_1b"].value_counts()

import matplotlib.pyplot as plt

labels = counts.index
sizes = counts.values

plt.figure(figsize=(18, 18))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Fruit Distribution")
plt.show()
```

```python
df = pd.read_csv("DataTUSA.csv")
counts = df["Genre_Over_1b"].value_counts()

import matplotlib.pyplot as plt

# Get the values and labels from the value_counts result
sizes = counts.values
labels = counts.index

# Make sure explode has the same length as the number of categories
# Adjust this based on how many unique values are in "Genre_Over_1b"
explode = [0.05] + [0] * (len(labels) - 1)  # First slice pulled out, others not

plt.pie(
    sizes, 
    labels=labels,          # category names on slices
    autopct="%1.1f%%",      # show percentages with 1 decimal place
    startangle=90,          # rotate so first slice starts at top
    shadow=True,            # optional: adds a drop shadow
    explode=explode         # dynamically sized explode list
)

plt.title("Genre Distribution")  # Changed title to match the data
plt.show()
```

Out of all of the genres present on Spotify's Top 50 Tracks USA, pop has the most songs over one billion.
