# Q: Which genre has the largest percentage of listens out of the total number of listens from every song on Spotify's Top 50 Tracks of the USA 2024?

I plan to conduct a observational study using a census to examine the whole population.

In order to gather this data, I used AI to generate a list of each song on the playlists along with their artists, Wikipedia genre, and AI genre, before going through each song and finding the number of listens. This data may produce a bias due to the fact that some of the data is pulled from Wikipedia, which is crowd-sourced.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TopUSAData.csv")
p_total = 0
c_total = 0
hh_total = 0
rb_total = 0
ap_total = 0
ar_total = 0
ir_total = 0
pr_total = 0
if_total = 0
total = 0
other = 0
for row in df.itertuples(index=True):
    if(row.Playlist == "USA"):
        if(row.AI_Genre == "Pop"):
            p_total = p_total + row.Num_of_Listens
        elif(row.AI_Genre == "Country"):
            c_total = c_total + row.Num_of_Listens
        elif(row.AI_Genre == "Hip-Hop"):
            hh_total = hh_total + row.Num_of_Listens
        elif(row.AI_Genre == "R&B"):
            rb_total = rb_total + row.Num_of_Listens
        elif(row.AI_Genre == "Alternative Pop"):
            ap_total = ap_total + row.Num_of_Listens
        elif(row.AI_Genre == "Alternative Rock"):
            ar_total = ar_total + row.Num_of_Listens
        elif(row.AI_Genre == "Indie Rock"):
            ir_total = ir_total + row.Num_of_Listens
        elif(row.AI_Genre == "Pop Rock"):
            pr_total = pr_total + row.Num_of_Listens
        elif(row.AI_Genre == "Indie Folk"):
            if_total = if_total + row.Num_of_Listens
        else:
            other = other + 1
total = p_total + c_total + hh_total + rb_total + ap_total + ar_total + ir_total + pr_total + if_total
print("Total number of listens on Top 50 USA:",  total)
print("Number of listens for Pop Songs:",  p_total)
print("Number of listens for Country Songs:", c_total)
print("Number of listens for Hip-Hop Songs:", hh_total)
print("Number of listens for R&B Songs:", rb_total)
print("Number of listens for Alternative Pop Songs:", ap_total)
print("Number of listens for Alternative Songs:", ar_total)
print("Number of listens for Indie Rock Songs:", ir_total)
print("Number of listens for Pop Rock Songs:", pr_total)
print("Number of listens for Indie Folk Songs:", if_total)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Added pandas import which was missing

# Example data
categories = ["Pop", "Country", "Hip-Hop", "R&B", "Alternative Pop", "Alternative Rock", "Indie Rock", "Pop Rock", "Indie Folk"]
values = [p_total, c_total, hh_total, rb_total, ap_total, ar_total, ir_total, pr_total, if_total ]

# Put into a dataframe (seaborn likes dataframes)
df = pd.DataFrame({"Genre": categories, "Number of Listens": values})

# Bar plot
sns.barplot(x="Genre", y="Number of Listens", data=df)  # Changed "Number of Listens" to "Listens" to match the dataframe column name

plt.title("Music Genre Listens")  # Updated title to match the data context
plt.xticks(rotation=45)  # Added rotation for better readability of genre names
plt.tight_layout()  # Added to prevent label cutoff
plt.show()
```
